```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x **2
```

```python
x
```

```python
y
```

```python
#funcional
plt.plot(x,y, "bo-")
#plt.show()
plt.xlabel("x labol")
plt.ylabel("y labol")
plt.title("Graph")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"ro-")

plt.subplot(1,2,2)
plt.plot(y,x,"bo-")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y, "bo-")
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig1 = plt.figure()

axes1 = fig1.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig1.add_axes([0.2, 0.4, 0.3, 0.4])

#ploting data of axes1
axes1.plot(x,y, "ro-")
axes1.set_xlabel("x label 1")
axes1.set_ylabel("y label 1")
axes1.set_title("Title 1")

# ploting data of axes2

axes2.plot(y,x, "bo-")
axes2.set_xlabel("x label 2")
axes2.set_ylabel("y label 2")
axes2.set_title("Title 2")
```
